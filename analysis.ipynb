# Visualization of the most popular product for each of the packages

```python
group2 = data.groupby("package_info")
```

**Popular products under the enterprise package**

```python
enterprise_group = group2.get_group("enterprise")
```

```python
enterprise_group
```

```python
enterprise_product_dist = enterprise_group["product_name"].value_counts()
```

```python
enterprise_product_dist[:40].plot(kind="barh", figsize = (8, 7))
plt.title("Most popular product under the Enterprise package")
```

# Prices of the most popular products under the enterprie package

```python
popular_pdts = enterprise_product_dist[:40].index
```

```python
popular_pdts_df = enterprise_group[enterprise_group["product_name"].isin(popular_pdts)]
```

```python
popular_pdts_prices = popular_pdts_df.groupby(["price", "product_name"]).count()
```

```python
popular_pdts_prices.plot(kind = "barh", figsize = (10, 30))
plt.title("Prices distribution of the most popular products")
```
